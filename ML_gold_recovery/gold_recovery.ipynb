# Проект: Восстановление золота из руды.

Автор: Семен Лимонов.

Задача — разработать прототип модели машинного обучения для эффективной работы промышленных предприятий. Модель должна предсказывать коэффициент восстановления золота из золотосодержащей руды на основе данных с параметрами входящей руды и показателями процесса флотации и очистки. Модель поможет оптимизировать производство, чтобы не запускать предприятие с убыточными характеристиками.

Метрика качества - sMAPE (англ. Symmetric Mean Absolute Percentage Error, «симметричное среднее абсолютное процентное отклонение») для оценки эффективности обогащения отдельно чернового (rougher) и финального (final) концентрата. Итоговая метрика рассчитывается по формуле: Итоговое sMAPE = 25% sMAPE (rougher) + 75% sMAPE (final).

Описание данных: 

Технологический процесс

-	Rougher feed — исходное сырье
-	Rougher additions (или reagent additions) — флотационные реагенты Xanthate, Sulphate, Depressant: 
 -	Xanthate — ксантогенат (промотер, или активатор флотации)
 -	Sulphate — сульфат (на данном производстве сульфид натрия)
 -	Depressant — депрессант (силикат натрия)

-	Rougher process (англ. «грубый процесс») — флотация
-	Rougher tails — отвальные хвосты
-	Float banks — флотационная установка
-	Cleaner process — очистка
-	Rougher Au — черновой концентрат золота
-	Final Au — финальный концентрат золота

Параметры этапов

-	air amount — объём воздуха
-	fluid levels — уровень жидкости
-	feed size — размер гранул сырья
-	feed rate — скорость подачи

Наименование признаков:
[этап].[тип_параметра].[название_параметра]

Пример: rougher.input.feed_ag

Возможные значения для блока [этап]:
-	`rougher` — флотация
-	`primary_cleaner` — первичная очистка
-	`secondary_cleaner` — вторичная очистка
-	`final` — финальные характеристики

Возможные значения для блока [тип_параметра]:
-	`input` — параметры сырья
-	`output` — параметры продукта
-	`state` — параметры, характеризующие текущее состояние этапа
-	`calculation` — расчётные характеристики

Данные содержатся в трёх файлах:
-	gold_recovery_train_new.csv — обучающая выборка
-	gold_recovery_test_new.csv — тестовая выборка
-	gold_recovery_full_new.csv — исходные данные

Планируемые этапы исследования:

1. Подготовка данных (открытие файла с данными, изучение общей информации, проверка расчета целевого показателя recovery, анализ различий обучающей и тестовой выборки, предобработка данных).

3. Выбор лучшей модели с подбором гиперпараметров с использованием обучающей выбрки методом кросс-валидации, проверка лучшей модели на тестовой выборке и проверка на адекватность (sanity check).  

4. Общий вывод.

### Проверка правильности расчета показателя "эффективность обогащения" (Recovery).  

Рассчитаем на обучающей выборке значения rougher_output_recovery и final_output_recovery. Найдем значение метрики MAE между вновь рассчитаными и имеющимися показателями.  

```python
# создадим функцию для расчета показателя recovery
def check_recovery (row):
    
    concentrate = row['rougher_output_au']
    feed = row['rougher_input_feed_au']
    tail = row['rougher_output_tail_au']
    
    recovery = (concentrate * (feed - tail)) * 100 / (feed * (concentrate - tail))
    
    return recovery
```

```python
# создадим набор данных с исходным рассчетным показателем recovery и столбец с вновь рассчитанным показателем
# выведем показатели распределения
recovery_train = pd.DataFrame(df_train['rougher_output_recovery'].copy())
recovery_train['rougher_output_recovery_check'] = df_train.apply(check_recovery, axis=1)
recovery_train.describe()
```

```python
# рассчитаем значение mae и выведем на экран
mae = mean_absolute_error(recovery_train['rougher_output_recovery'], recovery_train['rougher_output_recovery_check'])
print('Среднее абсолютное отклонение MAE: {:.15f}'.format(mae))
```

**Вывод**: Среднее абсолютное отклонение между имеющимся показателем `rougher_recovery` и рассчитанным для проверки составило величину, близкую к нулю. Кроме того, основные значения показателей распределенния (среднее, стандартное отклонение, мин, макс, квартили) также совпадают. Следовательно, можно сделать вывод о том, что данный показатель рассчитан в соответствии с представленной методикой.

```python
# создадим функцию для расчета показателя final recovery
def check_recovery_final (row):
    
    concentrate = row['final_output_au']
    feed = row['rougher_input_feed_au']
    tail = row['final_output_tail_au']

    recovery = (concentrate * (feed - tail)) * 100 / (feed * (concentrate - tail))
      
    return recovery
```

```python
# создадим набор данных с исходным рассчетным показателем recovery и столбец с рассчитанным отдельно показателем
# выведем показатели распределения

recovery_final = pd.DataFrame(df_train['final_output_recovery'].copy())
recovery_final['final_output_recovery_check'] = df_train.apply(check_recovery_final, axis=1)
recovery_final.describe()
```

```python
# рассчитаем значение mae и выведем на экран
mae = mean_absolute_error(recovery_final['final_output_recovery'], recovery_final['final_output_recovery_check'])
print('Среднее абсолютное отклонение MAE: {:.15f}'.format(mae))
```

**Вывод**: Среднее абсолютное отклонение между имеющимся показателем `final_recovery` и рассчитанным для проверки составило величину, близкую к нулю. Кроме того, основные значения показателей распределенния (среднее, стандартное отклонение, мин, макс, квартили) также совпадают. #### Масштабирование значений нормализацией.

Масштабируем значения обучающей и тестовой выборок нормализацией с помощью метода MinMaxScaler().

```python
# создадим функцию для масштабирования
def scale_X (X_train, X_test, columns):
    scaler = MinMaxScaler()
    scaler.fit(X_train)
    X_train[columns] = scaler.transform(X_train[columns])
    X_test[columns] = scaler.transform(X_test[columns])   
```

## Разработка прототипа модели машинного обучения.

### Создадим функцию для вычисления итоговой sMAPE.

```python
def calc_smape (y, y_pred):
    return 1/len(y) * sum(2 * abs(y-y_pred) / (abs(y) + abs(y_pred))) * 100
```

```python
def total_smape(smape_rough, smape_final):
    total_smape = smape_rough * 0.25 + smape_final * 0.75
    return total_smape
```

### Обучение разных моделей и оценка их качества кросс-валидацией. Выбор лучшей модели и проверка её на тестовой выборке.

```python
# создадим метрику для оценки моделей в процессе кросс-валидации
smape_score = make_scorer(calc_smape, greater_is_better=False)
```

```python
# выведем итоговые размеры выборок с признаками для проверки
print(X_train_rough.shape, X_train_final.shape, y_train_rough.shape, y_train_final.shape)
print(X_test_rough.shape, X_test_final.shape, y_test_rough.shape, y_test_final.shape)
```

```python
# создадим функцию для обучения моделей с помощью методов GridSearchCV и RandomizedSearchCV со встроенной оценкой 
# методом кросс-валидации

def validate_model (search, model, grid_param):
    
    if search != 'random':
        
        model_rough = model
        grid_rough = GridSearchCV(model_rough,
                                  param_grid=grid_param,
                                  scoring=smape_score,
                                  n_jobs=-1
                                 )
        grid_rough.fit(X_train_rough, y_train_rough)

        model_final = model
        grid_final = GridSearchCV(model_final,
                                  param_grid=grid_param,
                                  scoring=smape_score,
                                  n_jobs=-1
                                 )
        grid_final.fit(X_train_final, y_train_final)

        tot_smape = total_smape(- grid_rough.best_score_, - grid_final.best_score_)
        
    else:
        model_rough = model
        grid_rough = RandomizedSearchCV(model_rough,
                                        param_distributions=grid_param,
                                        scoring=smape_score,
                                        n_jobs=-1,
                                        random_state=12345
                                       )
        grid_rough.fit(X_train_rough, y_train_rough)

        model_final = model
        grid_final = RandomizedSearchCV(model_final,
                                        param_distributions=grid_param,
                                        scoring=smape_score,
                                        n_jobs=-1,
                                        random_state=12345
                                       )
        grid_final.fit(X_train_final, y_train_final)

        tot_smape = total_smape(- grid_rough.best_score_, - grid_final.best_score_)
            
    
    return grid_rough, grid_final, tot_smape
```

```python
%%time
# обучим и оценим молель LinearRegression
model_lr_rough, model_lr_final, best_lr_smape = validate_model('grid', LinearRegression(), {})
```

```python
%%time
# подберем гиперпараметры, обучим и оценим модель Lasso
model_lasso_rough, model_lasso_final, best_lasso_smape = validate_model('grid', Lasso(), 
                                                                        {'alpha': np.arange(0.01, 0.15, 0.01)})
```

```python
%%time
# подберем гиперпараметры, обучим и оценим молель DecisionTreeRegressor
model_dtr_rough, model_dtr_final, best_dtr_smape = validate_model('random', DecisionTreeRegressor(random_state=12345), 
                                                                        {'splitter': ('best', 'random'),
                                                                         'max_depth': range(1, 20),               
                                                                         'min_samples_split': range(2, 4),
                                                                         'min_samples_leaf': range(1, 3),
                                                                        }
                                                                       )
```

```python
%%time
# подберем гиперпараметры, обучим и оценим молель GradientBoostingRegressor
# время обучения около 7 минут
model_gbr_rough, model_gbr_final, best_gbr_smape = validate_model('random', GradientBoostingRegressor(random_state=12345),
                                                                  {'n_estimators': range(30, 80, 10),
                                                                   'max_depth': range(1, 5),
                                                                   'min_samples_split': range(2, 4),                              
                                                                   'min_samples_leaf': range(1, 3),
                                                                  }
                                                                 )
```

```python
# выведем результаты лучших значений итоговой метрики smape для исследованных моделей в процессе кросс-валидации
# и подбора гиперпараметров
results = pd.DataFrame({'Best total sMAPE':[best_lr_smape, best_lasso_smape, best_dtr_smape, best_gbr_smape]}, 
                       index=['LinearRegression', 'Lasso', 'DecisionTreeRegressor', 'GradientBoostingRegressor'])
results.sort_values(by='Best total sMAPE')
```

Вывод: лучшую итоговую метрику sMAPE продемонстрировала модель GradientBoostingRegressor, т.е. симметричная средняя процентная ошибка составила 7,74 % по шкале от 0 до 200 %.

Проверим модель на тестовой выборке и сравним с моделью DummyRegressor.

```python
# создадим функцию для проверки модели на тестовой выборке
def test_model (model_rough, model_final):
    y_pred_test_rough = model_rough.predict(X_test_rough)
    y_pred_test_final = model_final.predict(X_test_final)
    
    smape_r = calc_smape(y_test_rough, y_pred_test_rough)
    smape_f = calc_smape(y_test_final, y_pred_test_final)
    score = 0.25 * smape_r + 0.75 * smape_f
    return score
```

```python
# подберем лучшие гиперпарамтеры и обучим модель DummyRegressor
model_dummy_rough, model_dummy_final, best_dummy_smape = validate_model('grid', DummyRegressor(), 
                                                                        {'strategy':('mean', 'median', 'quantile'),
                                                                         'quantile': np.arange(0, 1, 0.05)
                                                                        }
                                                                       )
```

```python
# проверим модели GradientBoostingRegressor и DummyRegressor на тестовой выборке и выведем результат на экран
test_results = pd.DataFrame({'Best sMAPE':[best_gbr_smape, test_model(model_gbr_rough, model_gbr_final), test_model(model_dummy_rough, model_dummy_final)]}, 
                       index=['GradientBoostingRegressor cross-valid', 'GradientBoostingRegressor TEST', 'DummyRegressor TEST'])
```

```python
# выведем на экран результат проверки на тестовой выборке
display(test_results)
```

Вывод: Метрика лучшей модели GradientBoostingRegressor на тестовой выборке ухудшилась незначительно и составила 8,04 %, при этом оставаясь лучше метрики модели DummyRegressor. Таким образом, можно рекомендовать лучшую модель GradientBoostingRegressor для оценки целесообразности запуска производств.
